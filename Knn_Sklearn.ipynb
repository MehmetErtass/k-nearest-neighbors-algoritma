```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Iris.csv") # veri setinin yüklenmesi
```

```python
X = data.to_numpy() # X'i özellik veya giriş parametreleri olarak yapılması
X = X[:,1:5]
```

```python
Y = data.to_numpy() # Y'nin veri noktalarının veya özelliklerin çıktısını (etiketini) içerecek şekilde yapılmasıı
Y = Y[:,5]
```

```python
# Verilerin %80'i modeli eğitmek için kullanılacak şekilde bölmek ve Modları test etmek için %20 veri kullanılır
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.2) 
```

```python
test_accuracy = []  # Karşılık gelen K değerinin doğruluk değerlerini tutacak boş listedir
K = np.arange(1,34) # K bir listedir [1,2,.....34] e kadar
```

```python
# K değerleri üzerinde döngü 
for k in K: 
    knn = KNeighborsClassifier(n_neighbors=k) # Model oluşturma
    knn.fit(X_train, y_train)  # Modelin uyum
    # Hesaplama eğitimi ve veri doğruluğunu test etme 
    test_accuracy.append(knn.score(X_test, y_test)) # Tahmin etme ve doğruluğu hesaplama
```

```python
# Doğruluk grafiği ve K değeri
plt.plot(K, test_accuracy, label = 'Testing dataset Accuracy') 
plt.legend() 
plt.xlabel('n_neighbors') 
plt.ylabel('Accuracy') 
plt.show() 
```
