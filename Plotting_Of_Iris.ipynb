```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Iris.csv")  # veri setinin yüklenmesi
```

```python
setosa=data[data['Species']=='Iris-setosa']           # Setosa etiketli veri noktaları
versicolor=data[data['Species']=='Iris-versicolor']   # Etiket olarak Versicolor'a sahip veri noktaları
virginica =data[data['Species']=='Iris-virginica']    # Etiket olarak Verginica'ya sahip veri noktaları
```

```python
plt.figure()             
fig,ax=plt.subplots(1,1)  # Üç etiketi de çizmek için (aynı çizim)

setosa.plot(x="SepalLengthCm", y="SepalWidthCm", kind="scatter",ax=ax,label='setosa',color='r')
versicolor.plot(x="SepalLengthCm",y="SepalWidthCm",kind="scatter",ax=ax,label='versicolor',color='b')
virginica.plot(x="SepalLengthCm", y="SepalWidthCm", kind="scatter", ax=ax, label='virginica', color='g')
```
